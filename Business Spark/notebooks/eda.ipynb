```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Desktop/Business Spark/data/raw/train.csv")
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'],dayfirst=True)
data['Ship Date']=pd.to_datetime(data['Ship Date'],dayfirst=True)
```

```python
data=data.sort_values('Order Date')
data.head()
```

```python
data.isna().sum()
```

```python
data.isnull().sum()
```

```python
#We have 11 na/null values in postal code. Comparing with the total number of data thats fine for now. So we are currently negliging the data
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,5))
data['Sales'].plot()
plt.ylabel('Sales', fontsize=12)
plt.xlabel('Date', fontsize=12)
plt.title('Sales Details')
plt.show()
```

```python
monthly=data.set_index('Order Date').resample('M')['Sales'].sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly.index,monthly.values)
plt.ylabel('Sales', fontsize=12)
plt.xlabel('Month', fontsize=12)
plt.title('Sales Details of Each Month')
plt.show()
```

```python
#So this graph plots the monthly sales of the data. These spikes and dips shows the presence of seasonal changes in the sales.
```

```python
data.groupby('Region')['Sales'].sum()
```

```python
data.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```
